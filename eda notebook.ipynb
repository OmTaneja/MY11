Removing Duplicates

```python
def removing_duplicates(df,id_columns: list, sort_by = 'timestamp'):
    df=df.sort_values(by=sort_by,ascending=False)
    df=df.drop_duplicates(subset=id_columns,keep='first')
    return df
```

```python
def calculate_runs(df):
    for index, row in df.iterrows():
        if pd.isnull(row['runs']):
            if not pd.isnull(row['strike_rate']) and not pd.isnull(row['balls_faced']) and row['balls_faced'] != 0:
                # Calculate runs using strike_rate and balls_faced
                calculated_runs = (row['strike_rate'] * row['balls_faced']) / 100
                df.at[index, 'runs'] = round(calculated_runs, 0)  # Round to the nearest integer explicitly
            elif row['strike_rate'] == 0:
                # Replace runs null value with 0 if strike_rate is 0
                df.at[index, 'runs'] = 0
    return df
```

```python
import pandas as pd

def calculate_strike_rate(df):
    for index, row in df.iterrows():
        if pd.isnull(row['strike_rate']):
            if not pd.isnull(row['runs']) and not pd.isnull(row['balls_faced']) and row['balls_faced'] != 0:
                # Calculate strike_rate using runs and balls_faced
                calculated_strike_rate = (row['runs'] / row['balls_faced']) * 100
                df.at[index, 'strike_rate'] = round(calculated_strike_rate, 2)  # Round to two decimal places explicitly
            elif row['runs'] == 0:
                # Replace strike_rate null value with 0 if runs is 0
                df.at[index, 'strike_rate'] = 0
    return df
```

```python
def fill_missing_runs(df):

  for index, row in df.iterrows():
    if pd.isnull(row['runs']):  # Check for null value in 'runs'
      # Ensure valid values for calculation (not null and balls_faced > 0)
      if not pd.isnull(row['economy']) and not pd.isnull(row['balls_bowled']) :
        # Calculate estimated runs using economy and overs bowled
        estimated_runs = (row['economy'] * (row['balls_bowled'] / 6))  
        df.at[index, 'runs'] = round(estimated_runs,0)
  return df
```

```python
def fill_missing_economy(df):
    for index, row in df.iterrows():
        if pd.isnull(row['economy']):  # Check for null value in 'economy'
            # Ensure valid values for calculation (not null and balls_bowled > 0)
            if not pd.isnull(row['runs']) and not pd.isnull(row['balls_bowled']) and row['balls_bowled'] > 0:
                # Calculate estimated economy using runs and overs bowled
                estimated_economy = row['runs'] / (row['balls_bowled'] / 6)
                df.at[index, 'economy'] = round(estimated_economy, 2)
    return df
```

Identify the bowler id, total no of wickets taken, and number of unique venues for the bowler who took wickets at maximum number of unique venues, given
the matches were won by their team by wickets.

```python
import pandas as pd

# Assume match_df and bowler_df are preprocessed and already available
match_details = match_df
bowler_details = bowler_df

# Filter matches that were won by the team by wickets
won_by_wickets = match_details[match_details['by'] == 'wickets']

# Merge the filtered match details with bowler details on 'match_id'
merged_data = pd.merge(won_by_wickets, bowler_details, on='match id', how='inner')

# Group by bowler ID and calculate total wickets and unique venues
bowler_stats = merged_data.groupby('bowler_id').agg({
    'wicket_count': 'sum',
    'venue': pd.Series.nunique
}).reset_index()

# Identify the bowler with the maximum number of unique venues
top_bowler = bowler_stats.loc[bowler_stats['venue'].idxmax()]

# Output the required information
bowler_id = top_bowler['bowler_id']
total_wickets = top_bowler['wicket_count']
unique_venues = top_bowler['venue']

print(f"Bowler ID: {bowler_id}")
print(f"Total Wickets: {total_wickets}")
print(f"Number of Unique Venues: {unique_venues}")
```
